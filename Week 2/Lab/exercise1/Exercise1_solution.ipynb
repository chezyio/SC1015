Try changing 2016 in the URL to 2012 or 2008 or 2004 to see for yourself. This allows us to fetch the Olympics medal table from all these years (in fact, any year) quite easily. </p>

```python
base = "https://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table"

def updateYear(year):
    url = "https://en.wikipedia.org/wiki/" + str(year) + "_Summer_Olympics_medal_table"
    return url


for i in range(2000, 2017, 4):
    year = updateYear(i)
    data = pd.read_html(year)
    df = pd.DataFrame(data[2].head(20))
    display(df)
    
    
    
```
